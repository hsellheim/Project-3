I want only the results for the AfD. In the German voting system, you get two votes: One for the direct candidate, one for the party. Since the AfD doesn't have a direct candidate in all districts and I think the second votes are generally more representative for party approval (there is often lots of tactics involved when choosing which direct candidate to vote for), I want only those. From checking the columns, I can tell that "Unnamed: 38" must be second votes for the AfD but I'm double-checking to be sure.
