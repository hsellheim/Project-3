```python
from bs4 import BeautifulSoup
import requests

response = requests.get("https://www.colognepride.de/")
doc = BeautifulSoup(response.text)
```

```python
doc
```

```python
items = doc.find_all(class_='marquee _2')
for item in items:
    marquee_content = item.find(class_="marquee-content scroll")
    if marquee_content:
        links = marquee_content.select('a.w-inline-block:not(.hide)')
        for link in links:
            img = link.find('img')
            if img and img.get("src"):
              print(img["src"])
```
